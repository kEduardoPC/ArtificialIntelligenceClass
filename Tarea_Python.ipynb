```python
df = pd.DataFrame(Data)

df.head(40)
```

```python
print(df.describe())
```

```python
Measurement=[]
Length=[]

for i in range(len(df)-1):
  Measurement.append(df.Measurement[i])
  i+=1
print(Measurement)

for i in range(len(df)-1):
  Length.append(df.Length[i])
  i+=1
print(Length)
```

```python
from statistics import mean, median, mode, stdev, variance, quantiles

#Measurement Statistics
print("Measurement")
print("Mean=",mean(Measurement))
print("Median=",median(Measurement))
print("Mode=",mode(Measurement))
print("StDev=",stdev(Measurement))
print("Variance=",variance(Measurement))
print("Quantiles=",quantiles(Measurement))
```

```python
#Length Statistics
print("Length (m)")
print("Mean=",mean(Length))
print("Median=",median(Length))
print("Mode=",mode(Length))
print("StDev=",stdev(Length))
print("Variance=",variance(Length))
print("Quantiles=",quantiles(Length))
```
